```python
import wget
```

```python
homero = 'https://raw.githubusercontent.com/FernandaNavarro/VisualRecognition/master/homero.zip'
filehomero= wget.download(homero)
```

```python
marge = 'https://raw.githubusercontent.com/FernandaNavarro/VisualRecognition/master/marge.zip'
filemarge= wget.download(marge)

otro = 'https://raw.githubusercontent.com/FernandaNavarro/VisualRecognition/master/otro.zip'
fileotro= wget.download(otro)

desconocido = 'https://raw.githubusercontent.com/FernandaNavarro/VisualRecognition/master/desconocido.zip'
filedesconocido= wget.download(desconocido)

print(filehomero)
print(filemarge)
print(fileotro)
print(filedesconocido)
```

```python
with open(abspath('./homero.zip'), 'rb') as homero, \
     open(abspath('./otro.zip'), 'rb') as otro, \
     open(abspath('./marge.zip'), 'rb') as marge:
     classifier = service.create_classifier('TheSimpson',
                                            positive_examples={  'homero': homero, 
                                                                 'marge':marge},
                                            negative_examples=otro).get_result()
print(json.dumps(classifier, indent=2))
```

```python
#To run the classifier
classifiers = service.list_classifiers().get_result()
print(json.dumps(classifiers, indent=2))
```

```python
#with this you can run the classifier with some pictures
car_path = abspath('./desconocido.zip')
try:
    with open(car_path, 'rb') as images_file:
        simpson_results = service.classify(
            images_file=images_file,
            threshold='0.1',
            classifier_ids=['TheSimpson_802539214']).get_result()
        print(json.dumps(simpson_results, indent=2))
except ApiException as ex:
    print(ex)
```

```python
#To eliminate the classifier
response = service.delete_classifier(classifier_id='TheSimpson_1066577007').get_result()
print(json.dumps(response, indent=2))
```
